```python
#Importing required Libraries
import pandas as pd

#Loading the DataSet
data = pd.read_csv(r"C:\Users\rajin\OneDrive\Desktop\Internships\Oasis - Data Science\Task 4 - Spam E-Mail Detection\spam.csv",encoding = 'latin-1')
#specifying encoding information , because the emails contain different symbols which the pandas consider as utf-8 codec.
print(data.head())
print(data.info())
```

```python
#Finding Data Shape, Data columns and counts of 'spam' and 'not_spam"
print("Data Shape : \n",data.shape)
print("\nData Columns : \n",data.columns)
print("\n",data['label'].value_counts())
```

```python
#Since the unnamed columns provide no contribution to the model analysis, drop them
data = data[['label','email']]
print(data.head())
```

```python
#Mapping Values to labels:
data['label'] = data['label'].map({'not_spam':0, 'spam':1})
print (data.head())
```

TEXT PREPROCESSING

```python
#Importing required library
import re

#Pre-processing:
def preprocess_text(text):
    
    #converting text to lowercase
    text = text.lower()
    #removing urls
    text = re.sub(r'https\S+|www\S+|http\S+','',text)
    #removing non-alphabetic characters
    text = re.sub(r'[^a-z\s]','',text)

    return text

data['clean_email'] = data['email'].apply(preprocess_text)
print(data.head(10))
```

Vectorization 

```python
#Importing required libraries
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=3000, stop_words='english')

x = tfidf.fit_transform(data['clean_email'])
y = data['label']

print (x)
```

Train - Test Split and Naive Bayes Model

```python
#Importing libraries
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify = y, random_state=42)

model = MultinomialNB()
model.fit(x_train, y_train)

y_prediction = model.predict(x_test)

cm = confusion_matrix(y_test,y_prediction)
ac = accuracy_score(y_test,y_prediction)

print ("\nClassification Report :\n", classification_report(y_test, y_prediction))
print ("\nConfusion Matrix :\n",cm)
print ("\nAccuracy Score : \n", ac)
```

Predicting with Custom E-Mail :

```python
def predict_email(email):

    cleaned = preprocess_text(email)
    vectorized = tfidf.transform([cleaned])
    result = model.predict(vectorized)

    return "Spam Email !" if result[0] == 1 else "Not Spam Email :)"

#test
print(predict_email("Hi Nidhi. I hope your internship is going well. All the best"))
```
